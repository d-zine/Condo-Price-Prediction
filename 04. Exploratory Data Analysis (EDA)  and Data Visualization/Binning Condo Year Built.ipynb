```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import StrMethodFormatter
import matplotlib.ticker as ticker
import pandas as pd
```

```python
df=pd.read_csv('test2.csv')
```

```python
df['yearBuilt']=pd.cut(df.yearBuilt,bins = [0,1900,1980,2000,2023],labels=['pre-1900s','1900-1980s','1980-2000','2000+'])
```

```python
#bar chart
for x in ['yearBuilt']:
    fg=sns.catplot(data=df,x=x,y='price',kind='bar',estimator=np.median,ci=False)
    sns.set()
    for ax in fg.axes.flat:
        ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))
    plt.title(str(x)+' vs Median Condo Price')
    plt.xlabel(x)
    plt.ylabel("Median Condo Price (in dollars)")
    plt.tight_layout()
#So that axis show
    plt.show()
```
