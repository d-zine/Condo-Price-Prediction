```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import StrMethodFormatter
import matplotlib.ticker as ticker
import pandas as pd
```

```python
df=pd.read_csv('test2.csv')
```

```python
df['bathroom_bins']=pd.cut(df.bathrooms,bins = [0,1,2,3,12],labels=['1','2','3','4+'])
```

```python
df['bedroom_bins']=pd.cut(df.bedrooms,bins = [0,1,2,3,8],labels=['1','2','3','4+'])
```

```python
x_list=['bathroom_bins','bedroom_bins']
```

```python
fig, axes = plt.subplots(1, 2, figsize=(9, 5), sharey=True)
fig.suptitle('Bathrooms have a stronger influence on Mean Condo Price than Bedrooms.')
# Bulbasaur
sns.barplot(ax=axes[0], x=df.bathroom_bins, y=df.price,ci=None)
axes[0].set_title('Bathrooms vs Mean Condo Price')
axes[0].yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))

sns.barplot(ax=axes[1], x=df.bedroom_bins, y=df.price,ci=None)
axes[1].set_title('Bedrooms vs Mean Condo Price')
plt.show()
```
