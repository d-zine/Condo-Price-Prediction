```python
import pandas as pd
import requests
import json
import time
```

```python
# show all columns
pd.set_option('display.max_columns', None)
city = 'Manhattan'
state = 'NY'

search_str = city + ', ' + state
#search_str =state
print('Search string:', search_str)

#change city, state per day
```

```python
# get keys
rapid_api_key = 'XXX' 

#get property detail
prop_detail_list = []
# get data
page=1
while page<=20:

    url = "https://zillow-com1.p.rapidapi.com/propertyExtendedSearch"


    querystring = {"location": search_str,
               "home_type":"Condos","page":page}

    headers = {
    'x-rapidapi-host': "zillow-com1.p.rapidapi.com",
    'x-rapidapi-key': rapid_api_key
    }

    z_for_sale_resp = requests.request("GET", url, headers= headers, params=querystring)

# wait 1 sec based on limit
    time.sleep(1.5) #limit of requests per second based on pricing rules

# transform to json
    z_for_sale_resp_json = z_for_sale_resp.json()
    df_z_for_sale = pd.json_normalize(data=z_for_sale_resp_json['props'])
    print(df_z_for_sale)
    prop_detail_list.append(df_z_for_sale)
    print(len(prop_detail_list))
    print(page)
    page=page+1
    print('Finished')
```

```python
merged = pd.concat(prop_detail_list)
merged=merged.drop_duplicates(ignore_index=True)
```

```python
print('Num of rows:', len(merged))
print('Num of cols:', len(merged.columns))
merged.tail()
```

```python
#Monthly Mortgage Payments without Property Tax, Home Owner's Insurance 
#https://www.zillow.com/mortgage-calculator/
#assumes a 20% down payment
city['principal']=city['price']*.8
```

```python
#https://www.zillow.com/mortgage-rates/30-year-fixed/
city["mortgageRates.thirtyYearFixedRate"]=city["mortgageRates.thirtyYearFixedRate"].fillna(city["mortgageRates.thirtyYearFixedRate"].median())
city['monthlyHoaFee']=city['monthlyHoaFee'].fillna(0)
#city['resoFacts.taxAnnualAmount']=city['resoFacts.taxAnnualAmount'].fillna(0).squeeze()
#type(city['resoFacts.taxAnnualAmount'])
```

```python
#Mortgage Payment formula:P*r(1+r)^n / (((1+r)^(n))-1) 
#Assumes 30 year fixed rate mortgage
#thirty year fixed rate is an annual rate that needs to be converted to monthly number
#Then add HOA fees
#Then add home owners insurance

city['mortgageRates.thirtyYearFixedRate'] = pd.to_numeric(city['mortgageRates.thirtyYearFixedRate'])
city['monthlyHoaFee'] = pd.to_numeric(city['monthlyHoaFee'])
#city['resoFacts.taxAnnualAmount'] = pd.to_numeric(city['resoFacts.taxAnnualAmount'])
#city['monthlytax']=city['resoFacts.taxAnnualAmount']/12
```

```python
city['monthlyrate'] = city["mortgageRates.thirtyYearFixedRate"]/100/12
```

```python
city['numerator']=(city['principal']*city['monthlyrate'])*((1+city['monthlyrate'])**(12*30))
```

```python
city['denominator']=((1+city['monthlyrate'])**(12*30))-1
```

```python
city['mortgage']=(city['numerator']/city['denominator'])+city['monthlyHoaFee']+city['annualHomeownersInsurance']/12  
```
