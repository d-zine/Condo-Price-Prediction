```python
from sqlalchemy import create_engine
import pandas as pd
import os
import psycopg2
```

```python
philly_shootings=pd.read_csv("shooting-victims-data.csv")
philly_shootings['label']='philly crime'
print(philly_shootings)
philly_shootings['latitude']=philly_shootings['Latitude']
philly_shootings['longitude']=philly_shootings['Longitude']
```

```python
#Merge Condo files
philly = pd.read_csv("philly_zillow5.csv")
yonkers = pd.read_csv("zillow5yonkers_ny.csv")
arlington = pd.read_csv("zillow5Arlington_VA.csv")
baltimore = pd.read_csv("zillow5baltimore_md.csv")
brooklyn1 = pd.read_csv("zillow5Brooklyn1_NY.csv")
brooklyn2 = pd.read_csv("zillow5Brooklyn2_NY.csv")
brooklyn3 = pd.read_csv("zillow5Brooklyn3_NY.csv")
boston = pd.read_csv("zillow5Boston_MA.csv")
charlotte = pd.read_csv("zillow5Charlotte_NC.csv")
jerseycity = pd.read_csv("zillow5jersey_city_nj.csv")
dc = pd.read_csv("washington_dc_zillow5.csv")
columbus = pd.read_csv("zillow5Columbus_OH.csv")
pittsburgh = pd.read_csv("zillow5Pittsburgh_PA.csv")
queens1 = pd.read_csv("zillow5QueensNY.csv")
queens2 = pd.read_csv("zillow5Queens2NY.csv")
queens3 = pd.read_csv("zillow5Queens3_NY.csv")
queens4 = pd.read_csv("zillow5Queens4_NY.csv")
queens5 = pd.read_csv("zillow5Queens5_NY.csv")
ct = pd.read_csv("zillow5CT.csv")
alexandria=pd.read_csv("zillow5Alexandria_VA.csv")
statenisland=pd.read_csv("zillow5StatenIsland_NY.csv")
manhattan1=pd.read_csv("zillow5ManhattanNY.csv")
manhattan2=pd.read_csv("zillow5Manhattan2_NY.csv")
manhattan3=pd.read_csv("zillow5Manhattan3_NY.csv")
manhattan4=pd.read_csv("zillow5Manhattan4_NY.csv")
manhattan5=pd.read_csv("zillow5Manhattan5_NY.csv")
manhattan6=pd.read_csv("zillow5Manhattan6_NY.csv")
manhattan7=pd.read_csv("zillow5Manhattan7_NY.csv")

df_list = [philly, yonkers, arlington, baltimore, brooklyn1, brooklyn2, brooklyn3, boston, charlotte,
           jerseycity,dc, columbus,pittsburgh,queens1,queens2,queens3,queens4,queens5,ct,alexandria,
          statenisland,manhattan1,manhattan2,manhattan3,manhattan4,manhattan5,manhattan6, manhattan7,philly_shootings]

# Merge all the dataframes in df_list
appended_df = pd.concat(df_list)
```

```python
print('Num of rows:', len(appended_df))
print('Num of cols:', len(appended_df.columns))
appended_df.drop_duplicates().head()
```
